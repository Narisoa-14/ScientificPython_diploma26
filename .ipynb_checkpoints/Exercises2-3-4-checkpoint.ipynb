```python
import math

# Function to compute derivatives manually
def derivative(f, x, order=1, h=1e-5):
    """Compute the nth derivative of f at x using finite differences"""
    if order == 0:
        return f(x)
    else:
        return (derivative(f, x + h, order-1, h) - derivative(f, x - h, order-1, h)) / (2*h)

# Function to compute Taylor series
def taylor_series_manual(f, x, N):
    """Compute first N terms of the Taylor series of f around 0"""
    terms = []
    for k in range(N):
        f_k = derivative(f, 0, k)  # kth derivative at 0
        term = f_k * (x**k) / math.factorial(k)
        terms.append(term)
    return sum(terms), terms

# Example function
def f(x):
    return math.exp(x)  # e^x

x_val = 4
N_terms = 4

approx, terms_list = taylor_series_manual(f, x_val, N_terms)
print("Taylor series approximation:", approx)
print("Terms:", terms_list)


```

```python
def matmul_list(A, B):
    n = len(A[0])        # number of rows in A
    m = len(B[0])     # number of columns in B
    p = len(B[0])        # number of columns in A / rows in B

    C = [[0 for col in range(m)] for row in range(n)]  

  
    for i in range(n):       # row index of A
        for j in range(m):   # column index of B
            for k in range(p):  # sum index
                C[i][j] += A[i][k] * B[k][j]
    return C
   
A =[[1, 2, 3],[4, 5, 6], [7, 8, 9]]
B = [[9, 8, 7],[6, 5, 4],[3, 2, 1]]
A = np.flatten(A)
B = np.flatten(B)
```

```python
def matmul(A, B, M, N, P):
    C = [0] * (M * P)  # initialize flattened C
    for i in range(M):
        for j in range(P):
            sum = 0
            for k in range(N):
                sum += A[i*N + k] * B[k*P + j]
            C[i*P + j] = sum
    return C
#matmul_list([ [1, 2, 3],[4, 5, 6], [7, 8, 9]],[[9, 8, 7],[6, 5, 4],[3, 2, 1]])
```

#### **Exercise 5:**

Plot a rainbow. Use your imagination and google to make it beautiful.

```python
import numpy as np
import matplotlib.pyplot as plt

# Create x values
x = np.linspace(0, 3, 400)

# Rainbow colors
colors = ['red', 'orange', 'yellow', 'green', 'blue', 'indigo', 'violet']

# Plot sine waves with rainbow colors
plt.figure(figsize=(10,4))
for i, color in enumerate(colors):
    y = np.sin(x) + i   # shift each curve upward
    plt.plot(x, y, color=color, linewidth=8 + i, 
             linestyle='-', marker='o', markersize=28, alpha=0.9)

# Add title and labels
plt.title("Rainbow ", fontsize=15)
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.grid(True, alpha=0.3)
plt.show()
```

#### **Exercise 6:**

Plot a nice-looking 3-D torus. Do this 3 times with different styles. 

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Torus parameters
R = 2.0  # major radius
r = 0.7  # minor radius

u = np.linspace(0, 2*np.pi, 100)
v = np.linspace(0, 2*np.pi, 100)
u, v = np.meshgrid(u, v)

x = (R + r*np.cos(v)) * np.cos(u)
y = (R + r*np.cos(v)) * np.sin(u)
z = r * np.sin(v)

# Plot
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(x, y, z)
ax.set_title('Torus  Smooth Surface')
plt.show()
```

#### **Exercise 8:**

Integrate the function `x**2 + 10*np.sin(x)` on the interval `[-10,10]`.
Find a way to make a plot like in textbooks where you color the area under the curve that you integrate.
Now integrate the same function symbolically using `sympy`, then put borders into it and compare the results.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import quad

# Define the function
def f(x):
    return x**2 + 10*np.sin(x)

# Integration interval
a, b = -10, 10

# Numerical integration using quad
area, error = quad(f, a, b)
print("Numerical integral:", area)

# Plotting
x = np.linspace(a, b, 400)
y = f(x)

plt.figure(figsize=(8,5))
plt.plot(x, y, label="f(x) = x^2 + 10*sin(x)", color="blue")
plt.fill_between(x, y, where=[(xi>=a and xi<=b) for xi in x], color="black", alpha=1)
plt.axhline(0, color='red', linewidth=0.8)
plt.title("Area under f(x) from -10 to 10")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.legend()
plt.show()
```

```python
import sympy as sp

# Symbolic variable
x = sp.symbols('x')

# Symbolic function
f_sym = x**2 + 10*sp.sin(x)

# Symbolic indefinite integral
F = sp.integrate(f_sym, x)
print("Indefinite integral F(x):", F)

# Definite integral from -10 to 10
area_sym = sp.integrate(f_sym, (x, -10, 10))
print("Definite integral (symbolic):", area_sym.evalf())
```
